## 1. Imputation

```python
# Feature별 결측치 수 확인
data.isnull().sum()
```

```python
# 결측치를 제거한 후 데이터 개수 확인
data_1 = data.dropna()

print(f'결측치 제거 전 데이터 개수 : {len(data)}개')
print(f'결측치 제거 후 데이터 개수 : {len(data_1)}개')
```

```python
# (1) 결측치를 0으로 대체
data_2 = data.fillna(0)
data_2.head()
```

```python
# (2) 결측치를 Mean이나 Median으로 대체
# Categorical Feature인 'Cabin'은 Mean이나 Median이 존재하지 않으므로, 결측치가 대체되지 않음

data_3 = data.fillna(data.mean())
data_4 = data.fillna(data.median())

print(data_3.head())
print(data_4.head())
```

```python
# (3) 결측치를 가장 빈도가 높은 값으로 대체
print(data.Age.value_counts(), '\n')   # 21.0
print(data.Fare.value_counts(), '\n')   # 7.7500
print(data.Cabin.value_counts(), '\n')   # B57, B59, B63, B66
```

```python
data_5 = data.copy()
data_5.loc[data_5['Age'] != data_5['Age'], 'Age'] = 21.0   # np.nan != np.nan 활용하여 결측치 대체
data_5.loc[data_5['Fare'] != data_5['Fare'], 'Fare'] = 7.7500
data_5.loc[data_5['Cabin'] != data_5['Cabin'], 'Cabin'] = random.choice(['B57', 'B59', 'B63', 'B66'])
data_5.isnull().sum()   # 결측치 대체 완료
```

```python
data_5.head()
```

## 2. Indicator Variables

```python
data['Age'].describe()
```

```python
# 'Age' Feature의 요약통계를 바탕으로 'Oldness'라는 새로운 column 생성

oldness = []

for age in data.Age:
  if age < 21:
    oldness.append('young')
  elif 21 <= age < 39:
    oldness.append('middle')
  else:
    oldness.append('old')

data['Oldness'] = oldness
```

```python
data.head()
```

## 3. Scaling

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

mm_scaler = MinMaxScaler()
sd_scaler = StandardScaler()
```

```python
# MinMax Scaling (0~1)
mm_scaled_age = mm_scaler.fit_transform(data['Age'].values.reshape(-1,1))
data['mm_scaled_age'] = mm_scaled_age

# Standard Scaling (-1~1)
sd_scaled_age = sd_scaler.fit_transform(data['Age'].values.reshape(-1,1))
data['sd_scaled_age'] = sd_scaled_age
```

```python
data.head()
```

## 4. LabelEncoding

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
encoded = encoder.fit_transform(data['Embarked'])
data['Embarked_class'] = encoded
```
